```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time

```

```python
#setting covidtracking url & request
url = "https://covidtracking.com/api/v1/states/daily.json"
response = requests.get(url).json()
#turning .json response as a dataframe
covid_state_df = pd.DataFrame(response)
#list of all column headers to review
print(covid_state_df.columns.values)
```
