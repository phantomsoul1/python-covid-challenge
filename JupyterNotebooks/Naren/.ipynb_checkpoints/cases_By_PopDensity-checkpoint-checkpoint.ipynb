```python
import pandas as pd
import matplotlib.pyplot as plt
import requests
#import gmaps
#from config import gkey
#import us
from pprint import pprint
import scipy.stats as st
import numpy as np
from scipy.stats import linregress
```

```python
resource_path = "../../Resources"
all_by_state_path = f"{resource_path}/AllByState.csv"
all_by_state_daily_path = f"{resource_path}/AllByState-Daily.csv"
```

```python
all_by_state_df = pd.read_csv(all_by_state_path)
all_by_state_df.head()
```

```python
all_by_state_daily_df = pd.read_csv(all_by_state_daily_path)
all_by_state_daily_df.head()
```

```python
max_date = all_by_state_daily_df["Date"].max()
max_date
    
```

```python
max_date_df = all_by_state_daily_df.sort_values("Date", ascending = False)
max_date_df.head()
```

```python
grouped_state_name = max_date_df.groupby(["State Name"])
grouped_state_name.first()

# confirmed = grouped_state_name["Confirmed"].max()
# confirmed

population_density = grouped_state_name["Population Density"].sum()
population_density

cases = grouped_state_name["Confirmed"].sum()
cases

# pop = grouped_state_name["Population"].mean()
# pop

# pop_df = pd.DataFrame({"Population": pop})
# pop_max = pop_df.sort_values("Population", ascending = False)
# pop_max.head()
```

```python
pop_density_df = pd.DataFrame({"Population Density": population_density,
                                "Confirmed": cases})
pop_density_vs_cases = pop_density_df.sort_values("Population Density", ascending = False).head()
pop_density_vs_cases


pop_density_vs_cases_df = pop_density_df.sort_values("Population Density", ascending = False)
pop_density_vs_cases_df
```

```python
cases_by_pop_density = pop_density_df.sort_values("Confirmed", ascending = False).head()
cases_by_pop_density
```

```python
cases_by_pop_density_df = pop_density_df.sort_values("Confirmed", ascending = False)
cases_by_pop_density_df
```

```python
pop_density_vs_cases.plot(kind = "bar", alpha = 0.65, color = ("r", "g"), align = "center", figsize = (7, 5), legend = True)
plt.title("Population Density compared to Cases", fontsize= 14, fontweight = "bold")
plt.xlabel("States")
plt.xticks(rotation = 45)
plt.tight_layout()
plt.savefig("Population Density compared to Cases.png")
plt.show()
```

```python
cases_by_pop_density.plot(kind = "bar", alpha = 0.65, color = ("r", "g"), align = "center", figsize = (7, 5), legend = True)
plt.title("Cases compared to Population Density", fontsize= 14, fontweight = "bold")
plt.xlabel("States")
plt.xticks(rotation = 45)
plt.tight_layout()
plt.savefig("Cases compared to Population Density.png")
plt.show()
```

```python
max_date_df

avg_pop = max_date_df.groupby(["State Name"])["Population"].mean()
avg_cases = max_date_df.groupby(["State Name"])["Confirmed"].mean()

plt.scatter(avg_pop, avg_cases, marker = "o", facecolors = "red", edgecolors = "black")
plt.title("Average Population vs. Average Cases", fontsize = 16, fontweight = "bold", color = "r")
plt.xlabel("Population (1.0 x 10^7)")
plt.ylabel("Avg. Cases")
plt.grid()
plt.show()
```

```python
avg_pop_den= max_date_df.groupby(["State Name"])["Population Density"].mean()
avg_cases = max_date_df.groupby(["State Name"])["Confirmed"].mean()

plt.scatter(avg_pop_den, avg_cases, marker = "o", facecolors = "red", edgecolors = "black")
plt.title("Average Population Density vs. Average Cases", fontsize = 16, fontweight = "bold", color = "r")
plt.xlabel("Population Density")
plt.ylabel("Avg. Cases")
plt.grid()
plt.show()
```

```python
correlation = st.pearsonr(avg_pop_den, avg_cases)
print(correlation)
corr = (f"The correlation between both is {round(correlation[0], 2)}")
```

```python
x_values = avg_pop_den
y_values = avg_cases
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope, 2)) + "x +" + str(round(intercept, 2))
plt.scatter(x_values, y_values, marker = "o", facecolors = "red", edgecolors = "black")
plt.plot(x_values, regress_values, "r-")
plt.annotate(line_eq, (200, 10000), fontsize = 15, color = "red")
plt.annotate(corr, (200, 7000), fontsize = 13, color = "blue")
plt.grid()
plt.title("Avg. Cases vs. Avg. Population Density", fontsize = 14, fontweight = "bold")
plt.xlabel("Avg. Population Density")
plt.ylabel("Avg. Cases")
print(f"The r-squared value is: {rvalue**2}")
plt.tight_layout()
plt.savefig("Avg. Population Density vs. Avg. Cases per State.png")
plt.show()
```
