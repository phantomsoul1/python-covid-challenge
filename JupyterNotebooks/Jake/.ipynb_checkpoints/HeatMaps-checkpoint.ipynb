```python
import pandas as pd
import requests
import gmaps
from config import gkey

resource_path = "../../Resources"
all_by_state_daily_path = f"{resource_path}/AllByState-Daily.csv"
```

```python
all_by_state_daily_df = pd.read_csv(all_by_state_daily_path)
all_by_state_daily_df.head()
```

```python
max_date_df = all_by_state_daily_df.sort_values("Date", ascending = False)
max_date_df.head()
```

```python
grouped_state_name = max_date_df.groupby(["State Name"])
grouped_state_name

population_density = grouped_state_name["Population Density"].max()
population_density

lat = grouped_state_name["Lat"].max()
lat

lon = grouped_state_name["Lon"].max()
lon

deaths = grouped_state_name["Deaths"].max()
deaths

pop_density_df = pd.DataFrame({"Population Density": population_density,
                               "Deaths": deaths,
                               "Lat": lat,
                               "Lon": lon})
pop_density_vs_deaths = pop_density_df.sort_values("Population Density", ascending = False)
pop_density_vs_deaths

deaths_by_pop_density_df = pop_density_df.sort_values("Deaths", ascending = False)
deaths_by_pop_density_df
```

```python
gmaps.configure(gkey)
```

```python
# Store 'Lat' and 'Lng' into  locations 
locations = deaths_by_pop_density_df[["Lat", "Lon"]].astype(float)

# pop_density_df = pop_density_df.dropna()
population_densitys= deaths_by_pop_density_df["Population Density"].astype(float)
```

```python
# Create a poverty Heatmap layer
fig = gmaps.figure()

heat_layer = gmaps.heatmap_layer(locations, weights= population_densitys, 
                                 dissipating=False, max_intensity= .990,
                                 point_radius = 1.0)

fig.add_layer(heat_layer)

fig
```

```python
# Convert deaths to list
deaths = deaths_by_pop_density_df["Deaths"].tolist()
```

```python
death_layer = gmaps.symbol_layer(
    locations, fill_color='rgba(150, 0, 0, 0.4)',
    stroke_color='rgba(0, 150, 0, 0.4)', scale=2,
    info_box_content=[f"Death amount: {death}" for death in deaths]
)


fig = gmaps.figure()
fig.add_layer(death_layer)

fig
```

```python
# Create a combined map
fig = gmaps.figure()

fig.add_layer(heat_layer)
fig.add_layer(death_layer)

fig
```
