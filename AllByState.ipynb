```python
# Dependencies and Setup
import pandas as pd
import requests
import us
from pprint import pprint

resource_path = "Resources"
cases_by_state_path = f"{resource_path}/CasesByState-Daily.csv"
income_pop_path = f"{resource_path}/IncomePopByState.csv"
```

```python
cases_df = pd.read_csv(cases_by_state_path)
cases_df.head()
```

```python
cases_df.rename(columns={'State' : 'State Name', 'State Abbr' : 'State'}, inplace=True)
cases_df.head()
```

```python
demo_df = pd.read_csv(income_pop_path)
demo_df.head()
```

```python
print(demo_df.count())
print(len(demo_df['State'].unique()))
```

```python
all_df = pd.merge(cases_df, demo_df, on='State', how='inner', validate='many_to_one')
all_df.head(50)
```

```python
all_df.to_csv("Resources/AllByState-Daily.csv", index=False)
```
