```python
# Dependencies and Setup
#import matplotlib.pyplot as plt
import pandas as pd
#import numpy as np
import requests
#import time
#from scipy.stats import linregress

import us

# Import API key
#from api_keys import weather_api_key

# Incorporated citipy to determine city based on latitude and longitude
#from citipy import citipy

# pprint - useful for browsing api results
from pprint import pprint
```

```python
dates = ['2020-01-01', '2020-01-15', '2020-02-01', '2020-02-15',
         '2020-03-01', '2020-03-15', '2020-04-01', '2020-04-15',
         '2020-05-01', '2020-05-15', '2020-06-01', '2020-06-15',
         '2020-07-01', '2020-07-15']
url = 'https://covid-api.com/api/reports'
```

```python
#Test run
query_params = {'date' : '2020-07-15',
                'iso' : 'USA'}
response = requests.get(url, params=query_params)
print(response.url)
json = response.json()
#pprint(json)
data = json['data']

cases_df = pd.DataFrame(columns=['Active', 'Confirmed', 'Date', 'Deaths', 'Fatality Rate', 'Recovered', 'State'])

for datum in data:
    active = datum['active']
    confirmed = datum['confirmed']
    date = datum['date']
    deaths = datum['deaths']
    fatality_rate = datum['fatality_rate']
    recovered = datum['recovered']
    state = datum['region']['province']
    
    data_row = {
        'Active' : active,
        'Confirmed' : confirmed,
        'Date' : date,
        'Deaths' : deaths,
        'Fatality Rate' : fatality_rate,
        'Recovered' : recovered,
        'State' : state
    }
    
    cases_df = cases_df.append(data_row, ignore_index=True)
    
cases_df.head()
```

```python
cases_df = pd.DataFrame(columns=[
    'Active', 
    'Confirmed', 
    'Date', 
    'Deaths', 
    'Fatality Rate', 
    'Recovered', 
    'State'])

for date in dates:
    query_params = {'date' : date,
                    'iso' : 'USA'}
    response = requests.get(url, params=query_params)
    json = response.json()
    data = json['data']

    for datum in data:
        active = datum['active']
        confirmed = datum['confirmed']
        date = datum['date']
        deaths = datum['deaths']
        fatality_rate = datum['fatality_rate']
        recovered = datum['recovered']
        state = datum['region']['province']

        data_row = {
            'Active' : active,
            'Confirmed' : confirmed,
            'Date' : date,
            'Deaths' : deaths,
            'Fatality Rate' : fatality_rate,
            'Recovered' : recovered,
            'State' : state
        }

        cases_df = cases_df.append(data_row, ignore_index=True)

cases_df.head()
```

```python
print(len(cases_df['State'].unique()))
```

```python
#state_names = []
#for state in us.states.STATES:
#    state_names.append(state.name)
    
clean_cases_df = cases_df[cases_df["State"].isin([state.name for state in us.states.STATES])]
clean_cases_df
```

```python
clean_cases_df.to_csv("Resources/CasesByState.csv", index=False)
```
