```python
url = "https://covid-19-statistics.p.rapidapi.com/reports"

# creates a list for storing information
province = []
confirmed = []

states = ["Alabama", "Alaska", "Arizona", "Arkansas", "California", "Colorado", "Connecticut", "Delaware", "Florida",
          "Georgia", "Hawaii", "Idaho", "Illinois", "Indiana", "Iowa", "Kansas", "Kentucky", "Louisiana", "Maine", 
          "Maryland", "Massachusetts", "Michigan", "Minnesota", "Mississippi", "Missouri", "Montana", "Nebraska",
          "Nevada", "New Hampshire", "New Jersey", "New Mexico", "New York", "North Carolina", "North Dakota", "Ohio",
          "Oklahoma", "Oregon", "Pennsylvania", "Rhode Island", "South Carolina", "South Dakota", "Tennessee", "Texas",
          "Utah", "Vermont", "Virginia", "Washington", "West Virginia", "Wisconsin", "Wyoming"]

# variable (x) for tracking state 
x = 0

for state in states:
    
    try:
        querystring = {"iso":"USA", "region_province": state}

        headers = {
        'x-rapidapi-host': "covid-19-statistics.p.rapidapi.com",
        'x-rapidapi-key': api_key
        }

        response = requests.get(url, headers=headers, params=querystring)
        response_json = response.json()
#         print(response.url)

        data = response_json['data']
#         pprint(data)
        s = data[0]['region']['province']
        c = data[0]['confirmed']

   
         # appends the data that has been processed
        confirmed.append(c)
        province.append(s)

        x += 1

#     # shows if a state/other data fails to be processed
    except (KeyError,IndexError):
        print("not found")
        x += 1

        
# create a dataframe and include all the data you want from the JSON file      
data_1 = {'date': '2020-07-01', 
          'confirmed': confirmed, 
          'province': province}

df = pd.DataFrame(data_1)
# # df.head()

# create a second dataframe to only include states in the United States and not PuertoRico/cruise ships
c_df = df[df['province'].isin([state for state in states])]
c_df.head()
    
```
